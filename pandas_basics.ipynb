```python
bios_new = bios.copy()
```

```python
bios_new['firts_name'] = bios_new['name'].str.split(' ').str[0]
```

```python
bios_new.head()
```

```python
bios_new['born_datetime'] = pd.to_datetime(bios_new['born_date'])
```

```python
bios_new['born_year'] = bios_new['born_datetime'].dt.year
bios_new[['name','born_year']]
```

```python
bios_new.to_csv('./data/new_bios.csv')
```

```python
bios_new['height_category'] = bios_new['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Average' if x < 185 else 'Tall'))
bios_new.head()
```

```python
def categoraize_athelete(row):
    if row['height_cm']  < 175 and row['weight_kg'] <70:
        return 'Lightweight'
    elif row['height_cm']  < 185 and row['weight_kg'] <80:
        return 'Middleweight'
    else:
        return 'Heavywight'
    
bios_new['Category'] = bios_new.apply(categoraize_athelete, axis=1)
```
